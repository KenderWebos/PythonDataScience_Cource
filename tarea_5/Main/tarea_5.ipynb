```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn import metrics
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
path = '../dataframe/fifa_procesado.csv'
df = pd.read_csv(path)
len(df)
```

```python
df
```

```python
df = df[df['Work Rate Defense'].notna()]
df = df[df['Position (4)'].notna()]
```

```python
#variables independientes
#['LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB']
datos = df.drop([ 'Position (13)','Name','Photo','Nationality','Flag','Club','Club Logo','Preferred Foot','Work Rate','Body Type','Real Face','Work Rate Attack', 'Position','LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB'],axis=1)
datos
```

```python
datos['Work Rate Defense'].unique()
```

```python
wrd = pd.get_dummies(datos["Work Rate Defense"])

# combinar los dataframes "df" y "variable_dummy_1" 
datos = pd.concat([datos, wrd], axis=1)

# Elimina la columna original "fuel-type" del dataframe "df"
datos.drop("Work Rate Defense", axis = 1, inplace=True)
datos
```

```python
datos.to_csv('../dataframe/fifa_datos_limpios.csv')
```
