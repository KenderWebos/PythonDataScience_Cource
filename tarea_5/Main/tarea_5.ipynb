## Cargamos las librerias a utilizar

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn import metrics
from sklearn.model_selection import train_test_split
%matplotlib inline
```

## Se carga el data set

```python
path = '../dataframe/fifa_procesado.csv'
df = pd.read_csv(path)
len(df)
```

## Vemos graficamente las tablas

```python
df
```

## Limpiamos el dataset de datos nulos y clomnuas que no seran consideradas en esete analisis

Limpiamos las columnas Work Rate Defense y Position (4) de filas con valores nulos

```python
df = df[df['Work Rate Defense'].notna()]
df = df[df['Position (4)'].notna()]
```

Un paso importante es definir las variable **X** e **y** y que atributos o características la van a componer. Para ello se procede a visualizar las columnas del conjunto de datos.

```python
print(df.columns)
```

Eliminamos columnas que no son relevantes para el analisis de clasificación 

```python
#variables independientes
#['LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB']
datos = df.drop(['Position (13)','Name','Photo','Nationality','Flag','Club','Club Logo','Preferred Foot','Work Rate','Body Type','Real Face','Work Rate Attack', 'Position','LS','ST','RS','LW','LF','CF','RF','RW','LAM','CAM','RAM','LM','LCM','CM','RCM','RM','LWB','LDM','CDM','RDM','RWB','LB','LCB','CB','RCB','RB', 'Loaned From'],axis=1)
# Reemplazamos todos los valores nan por ceros 
datos = datos.fillna(0)
datos.head()
```

Cambiamos el formato de la columna Work Rate Defense, utilizando el metodo get_dummies de pandas para volverla numerica

```python
datos['Work Rate Defense'].unique()
```

```python
wrd = pd.get_dummies(datos["Work Rate Defense"])

# combinar los dataframes "df" y "variable_dummy_1" 
datos = pd.concat([datos, wrd], axis=1)

# Elimina la columna original "fuel-type" del dataframe "df"
datos.drop("Work Rate Defense", axis = 1, inplace=True)
datos.head()
```

teniendo listo el datatset para el analisis creamos un nuevo csv del dataset de fifa

```python
datos.to_csv('../dataframe/fifa_datos_limpios.csv')
```

Cargamos el dataset limpio para el analisis

```python
path = '../dataframe/fifa_datos_limpios.csv'
df_entrenamiento_0 = pd.read_csv(path)
df_entrenamiento = df_entrenamiento_0.drop(['Unnamed: 0'], axis=1)
df_entrenamiento.head()
```

Obserbamos informacion sobre las columnas que utilizaremos

```python
print(df_entrenamiento.info())
```

Tambien vemos cuántos hay en cada clase en el conjunto de datos.

```python
df_entrenamiento['Position (4)'].value_counts()
```

Hay 6838 jugadores en posición middle, 5866 jugadores en posición de defense, 3418 jugadores en posición de attack y 2025 jugadores en posición goalkeeper.

Se crea la variable df_y que contiene las posiciones de los jugadores y es la variable dependiente para el analisis

```python
df_y = df_entrenamiento['Position (4)']
df_y
```

Se define los valores de la variables independientes para el analisis 

```python
df_x = df_entrenamiento.drop(['Position (4)'],axis=1)
```

```python
X = df_x.values
X[0:5]
```

```python
y = df_y
y[0:5]
```

Creamos la data de entrenamiento y de pruebas para ver el rendimiento del modelo

```python
X3_entrena, X3_prueba, y3_entrena, y3_prueba = train_test_split(X, y, test_size=0.3, random_state=3)
print ('Conjunto de Entrenamiento set:', X3_entrena.shape,  y3_entrena.shape)
print ('Conjunto de Prueba:', X3_prueba.shape,  y3_prueba.shape)
```

## Preparacion del modelo
El modelo a utlizar es el de Arboles de Decisión

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
arbol_modelo = DecisionTreeClassifier(criterion="entropy", max_depth = 6)
arbol_modelo
```

Entrenamos el medelo con la data de entrenamiento

```python
arbol_modelo.fit(X3_entrena,y3_entrena)
```

## Pronóstico

Probamos el modelo con el set de prueba y lo almacenamos para hacer la evaluación 

```python
arbol_pronostico = arbol_modelo.predict(X3_prueba)
```

A simple vista se puede ver que para algunas clasificaciones acierta y en otras no.

```python
print (arbol_pronostico [0:6])
print (y3_prueba [0:6])
```

## Evalaución del modelo
Utilizando las métricas de **sklearn** y verifiquemos la precisión del modelo.

```python
print("Precición del modelo basado en árbol de decisiones: ", metrics.accuracy_score(y3_prueba, arbol_pronostico))
```

Como se puede ver el modelo tiene una precición sobre el 80% lo que nos da un indicador bastante realista de la clasificacion
